```python
from bs4 import BeautifulSoup 
from datetime import date
import requests 

# get filecoin blog site
html_text = requests. get('https://filecoin.io/blog/#posts').text 

soup = BeautifulSoup(html_text ,'lxml')

posts = soup.find_all('div', class_ ='div blog-card')

today = datetime.date.today()

for post in posts: 
    published_date = post.find('p', class_ = 'p author-date type-mono').text
    if str(today.year) not in published_date and str(today.year - 1) not in published_date:
        pass
    elif len(published_date) > 100:
        pass
    else:
        print(published_date.strip())
```

# Review Scrapper 

```python
import requests
from bs4 import BeautifulSoup

url = "https://myanimelist.net/anime/5114/Fullmetal_Alchemist__Brotherhood/reviews"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

reviews = soup.find_all("div", {"class": "text"})
review_texts = [review.text.strip() for review in reviews]

for review_text in review_texts:
    print(review_text)
```

```python
import requests
from bs4 import BeautifulSoup
import csv

base_url = "https://myanimelist.net/anime/5114/Fullmetal_Alchemist__Brotherhood/reviews"
review_texts = []

for i in range(1, 11):
    url = f"{base_url}?p={i}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    reviews = soup.find_all("div", {"class": "text"})
    review_texts += [review.text.strip() for review in reviews]

for review_text in review_texts:
    print(review_text)

with open("reviews.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    for review_text in review_texts:
        writer.writerow([review_text])
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://myanimelist.net/anime/5114/Fullmetal_Alchemist__Brotherhood/reviews"
review_texts = []

for i in range(1, 11):
    url = f"{base_url}?p={i}"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    reviews = soup.find_all("div", {"class": "text"})
    review_texts += [review.text.strip() for review in reviews]

for review_text in review_texts:
    print(review_text)

df = pd.DataFrame({"review_text": review_texts})
df.to_csv("reviews.csv", index=False, encoding="utf-8")
```

# Forum ID Scrapper

```python
import requests
from bs4 import BeautifulSoup

url = 'https://myanimelist.net/forum/?animeid=269&topic=episode&show=0'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
rows = soup.find_all('tr', {'id': lambda x: x and x.startswith('topicRow')})

t_id = []
for row in rows:
    t_id.append(row['data-topic-id'])
    
print(t_id)
```

```python
import requests
from bs4 import BeautifulSoup

def get_forum_ids(anime_id):
    all_t_id = []
    shows = 0
    page_num = 1
    while True:
        # construct the URL for the current page number
        url = f"https://myanimelist.net/forum/?animeid={anime_id}&topic=episode&show={shows}"
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        rows = soup.find_all('tr', {'id': lambda x: x and x.startswith('topicRow')})
        if not rows:
            break
        
        t_id = []
        for row in rows:
            t_id.append(row['data-topic-id'])
        print(f"Forum IDs for page {page_num}: {t_id}")
        all_t_id.extend(t_id)
        shows += 50
        page_num += 1
        
    #print(all_t_id)
get_forum_ids(269)
```
